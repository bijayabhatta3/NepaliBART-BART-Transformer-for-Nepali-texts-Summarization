```python
from transformers import BartForConditionalGeneration, AutoModelForSeq2SeqLM


model = AutoModelForSeq2SeqLM.from_pretrained("Bijayab/ekantipur90_nep_summ_BART",
                                              max_length = 1000,
                                              do_sample=True,
                                              num_beams = 3,
                                              )
```

```python
inputs = fast_tokenizer(sentence, max_length=1000, return_tensors="pt")
```

# Test Summary Generation

```python
import pandas as pd
from tqdm import tqdm
```

```python
df = pd.read_csv('/content/test_prediction.csv')
```

```python
def apply_summarization(text):
  inputs = fast_tokenizer(text, max_length=1000, return_tensors="pt")
  summary_ids = model.to('cuda').generate(inputs["input_ids"].to('cuda'), min_length=0, max_length=256, num_beams = 3, do_sample = True)
  return fast_tokenizer.decode(summary_ids[0], skip_special_tokens=True, clean_up_tokenization_spaces=False)
```

```python
tqdm.pandas(desc="Applying function")
df['summary_csv_file'] = df['text'].progress_apply(apply_summarization)
```
